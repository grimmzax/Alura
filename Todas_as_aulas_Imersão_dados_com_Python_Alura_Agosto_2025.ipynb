```python
df.columns = ['ano', 'senioridade', 'contrato', 'cargo',
               'salario', 'moeda', 'usd', 'residencia',
               'remoto', 'empresa', 'tamanho_empresa']
display(df.head())
```

```python
df["contrato"].value_counts()
```

```python
traducao_senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo',
    'Médio': 'Pleno',  # Add mapping for already translated 'Médio'
    'Entrada': 'Junior' # Add mapping for already translated 'Entrada'
}

df['senioridade'] = df['senioridade'].replace(traducao_senioridade)
display(df.head())
print(df['senioridade'].unique())
```

```python
print(df['senioridade'].unique())
```

```python
traducao_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(traducao_contrato)
display(df.head())
```

```python
traducao_tamanho_empresa = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(traducao_tamanho_empresa)
display(df.head())
```

```python
traducao_remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(traducao_remoto)
display(df.head())
```

```python
import numpy as np

# Criação de um dataframe de tste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]

})

#calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''calcula medina e substitui os nulos pela mediana'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, "Belém"],
})
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna("Não informado")
df_cidades
```

```python
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']

salario_medio_pais_ds = df_ds.groupby('empresa')['usd'].mean().reset_index()
salario_medio_pais_ds = salario_medio_pais_ds.sort_values('usd', ascending=False)

fig = px.bar(salario_medio_pais_ds, x='empresa', y='usd',
             title='Salário Médio para Cientistas de Dados por País',
             labels={'empresa': 'País', 'usd': 'Salário Médio Anual (USD)'})
fig.show()
```
